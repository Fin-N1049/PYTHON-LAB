```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

# 1. Overview of the dataset
print("Dataset Shape:", data.shape)
print("\nFirst 5 Rows of the Dataset:\n", data.head())
print("\nSummary of Dataset:\n", data.info())
print("\nStatistical Summary:\n", data.describe())

# 2. Check for missing values
print("\nMissing Values Before Imputation:\n", data.isnull().sum())

# 3. Handle missing values
# Replace missing values in the 'Fare' column with the mean
data['Fare'].fillna(data['Fare'].mean(), inplace=True)

print("\nMissing Values After Imputation:\n", data.isnull().sum())

# 4. Visualizing missing data (after imputation)
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap (After Imputation)")
plt.show()

# 5. Analyze numerical features
numerical_features = ['Age', 'Fare', 'SibSp', 'Parch']
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[feature], kde=True, bins=30, color='blue')
    plt.title(f"Distribution of {feature}")
    plt.show()

# 6. Analyze categorical features
categorical_features = ['Sex', 'Pclass', 'Embarked', 'Survived']
for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=feature, data=data, palette='Set2')
    plt.title(f"Countplot of {feature}")
    plt.show()

# 7. Analyze relationships
# Correlation matrix for numerical variables
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pairplot for numerical relationships (optional)
sns.pairplot(data, hue='Survived', vars=numerical_features, diag_kind='kde')
plt.show()

# 8. Analyze survival rate based on categorical features
for feature in ['Sex', 'Pclass', 'Embarked']:
    survived_rates = data.groupby(feature)['Survived'].mean()
    print(f"\nSurvival Rates by {feature}:\n", survived_rates)
    survived_rates.plot(kind='bar', figsize=(8, 4), color='skyblue', edgecolor='black')
    plt.title(f"Survival Rates by {feature}")
    plt.ylabel("Survival Rate")
    plt.show()
```
